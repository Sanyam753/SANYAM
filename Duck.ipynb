```python
import cv2
from google.colab.patches import cv2_imshow
img = cv2.imread('/content/ss.jpg')
cv2_imshow(img)
gr = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
invert = cv2.bitwise_not(gr)
blur = cv2.GaussianBlur(invert, (21, 21), 0)
ib = cv2.bitwise_not(blur)
sketch = cv2.divide(gr, ib, scale=256.0)
cv2_imshow(img)
cv2_imshow(sketch)
cv2.waitKey(0)
```
